```python
import pandas
```

```python
#For Classification : Supervised(Features(numbers), Outcome(text))
# Label Encoding: Process 
data = pandas.read_csv('bank_full.csv')
data
```

```python
data.isnull().sum()
```

```python
data['job'].fillna('Unknown', inplace=True)
data['marital'].fillna('Unknown', inplace=True)
data['duration'].fillna(0.0, inplace=True)
data.isnull().sum()
```

```python
data['loan'].fillna('Unknown',inplace=True)
data.isnull().sum()
```

```python
data[['job','marital','loan']].head(20)
```

```python
data.drop_duplicates
```

```python
data.head(30)
```

```python
data['poutcome'].replace({
    'failure':1,
    'nonexistent':2, 
    'success':3
},inplace=True)
data.groupby('poutcome').size()
```

```python
data['day_week'].replace({
    'fri':1,
    'mon':2, 
    'thu':3,
    'tue':4,
    'wed':5
},inplace=True)
data.groupby('day_week').size()
```

```python
data['month'].replace({
    'apr':1,
    'aug':2, 
    'dec':3,
    'jul':4,
    'jun':5,
    'mar':6,
    'may':7,
    'nov':8,
    'oct':9,
    'sep':10,
},inplace=True)
data.groupby('month').size()
```

```python
data['contact'].replace({
    'cellular':1,
    'telephone':2,
},inplace=True)
data.groupby('contact').size()
```

```python
data['marital'].replace({
    'Unknown':'unknown'
},inplace=True)
data.groupby('marital').size()
```

```python
data['marital'].replace({
    'divorced':1,
    'married':2,
    'single':3,
    'unknown':4
},inplace=True)
data.groupby('marital').size()
```

```python
data['job'].replace({
    'Unknown':'unknown'
},inplace=True)
data.groupby('job').size()
```

```python
# Encoding
# In python we use replace() to encode
data['job'].replace({
    'admin.':1,
    'blue+AC0-collar' :2,
    'entrepreneur':3,
    'housemaid':4,
    'management':5,
    'retired':6,
    'self+AC0-employed':7,
    'services':8,
    'student':9,
    'technician':10,
    'unemployed':11,
    'unknown':12
},inplace=True)
data.groupby('job').size()
```

```python
# education encoding
data['education'].replace({
    'basic.4y':1,
    'basic.6y':2,
    'basic.9y':3,
    'high.school':4,
    'illiterate':5,
    'professional.course':6,
    'university.degree':7,
    'unknown':8,
},inplace=True)
data.groupby('education').size()
```

```python
# default encoding
data['default'].replace({
    'no':1,
    'unknown':2,
    'yes':3,
},inplace=True)
data.groupby('default').size()
```

```python
# housing encoding
data['housing'].replace({
    'no':1,
    'unknown':2,
    'yes':3,
},inplace=True)
data.groupby('housing').size()
```

```python
data['loan'].replace({
    'Unknown':'unknown',
},inplace=True)
data.groupby('loan').size()
```

```python
# loan encoding
data['loan'].replace({
    'no':1,
    'unknown':2,
    'yes':3,
},inplace=True)
data.groupby('loan').size()
```

```python
data.drop(['duration'],axis=1,inplace=True)
```

```python
data.to_csv('BankCleaned1.csv')
data.isnull().sum()
```
