```python
import pandas as pd
data = pd.read_csv('cleanedDataset.csv')
data.head(5)
```

```python
# 1. What is the Average Study Time
mean_study = data['StudyTime'].mean()
mean_study
```

```python
# 2. What is the Average for Writing
mean_writing = data['Writing'].mean()
mean_writing
```

```python
# 3. What is the Max marks for Maths
max_math = data['Math'].max()
max_math
```

```python
# 4. What is the minimum height
min_height = data['Height'].min()
min_height
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.available
```

```python
# 5. Create a Countplot of LiveOnCampus
plt.style.use('ggplot')

sns.countplot(data['LiveOnCampus'],palette='Reds')
plt.xlabel('Number of Students')
plt.ylabel('Live On Campus')
```

```python
# 6. Create a Count plot of Rank
sns.countplot(data['Rank'],palette='Oranges')
plt.xlabel('Number of Students')
```

```python
average = data.groupby(['LiveOnCampus','HowCommute'])['SleepTime'].mean()
average
```

```python
# 7. Is there a difference in average sleep time between students who live on campus and those who commute?
import numpy
sns.barplot(x='LiveOnCampus', y='SleepTime',data=data,palette='Reds',estimator=numpy.mean)
sns.barplot(x='HowCommute', y='SleepTime',data=data,palette='Reds',estimator=numpy.mean)
```

```python
# 8. What proportion of students are ranked?(use bargraph for this)
sns.barplot(data['Rank'],palette='Oranges')
plt.title('Distribution of Ranks')
plt.xlabel('Number of Students')
plt.ylabel('Ranks')
```

```python
# 9. Create a Swarm Plot of Athlete and Writing, Include Gender (hue)
sns.swarmplot(x='Athlete',y='Writing',hue='Gender',palette='magma',data=data)
```

```python
# 10. How many students smoke?
ax = sns.countplot(x=data['Smoking'],palette='Reds')
plt.xlabel('Smoking')
plt.ylabel('Number of Students')
for p in ax.patches:
    txt = str(p.get_height().round(1)) + ''
    txt_x = p.get_x()
    txt_y = p.get_height()
    ax.text(txt_x,txt_y,txt)
```

```python
# 11. What are the 25th, 50th, and 75th percentiles of the number of students in the dataset?
data.describe()
```

```python
# 12. What is the distribution of commute times? Are there any outliers?(use histogram)
sns.histplot(data['CommuteTime'],palette='Oranges')
plt.title('Distribution of commute time')
plt.xlabel('Commute Time')
plt.ylabel('Number of Students')
```

```python
# 13. Is there a significant difference in average height between male and female students?
average = data.groupby('Gender')[['Height']].mean()
average
```

```python
# 14. How do average exam scores (English, Reading, Math, Writing) vary across different states?
average = data.groupby('State')[['English', 'Reading', 'Math', 'Writing']].mean()
average
```

```python
# 15. What is the average height and weight of the students?
mean_height = data['Height'].mean()
print(f'The mean height is {mean_height}')
mean_weight = data['Weight'].mean()
print(f'The mean weight is {mean_weight}')
```
