```python
# cleaning quantitative data types/columns
data.isnull().sum() # if cell is empty it will print true, if not empty it will print false
```

```python
#fillna()   it fills empty cells with a given value
data['Rank'].fillna(0.0, inplace=True)
data['Gender'].fillna(2.0, inplace=True)
data['Major'].fillna('unknown', inplace=True)
data['Smoking'].fillna(3.0, inplace=True)
data['State'].fillna('unknown', inplace=True)
data['LiveOnCampus'].fillna( 2.0, inplace=True)
data['HowCommute'].fillna(4.0 , inplace=True)
data.isnull().sum()


```

```python
# replacing cells with different values, replace()
data['Gender'].replace({
    0:'Male',
    1:'Female',
    2:'Neutral'
},inplace=True)
data['LiveOnCampus'].replace({
    0:'Off Campus',
    1:'On Campus',
    2:'unknown'
},inplace=True)
data['HowCommute'].replace({
    1:'Walk',
    2:'Bike',
    3:'Car',
    4:'Public Transit'
},inplace=True)
data['Rank'].replace({
    0:'Unknown',
    1:'First position',
    2:'First runners up',
    3:'Second runners up',
    4:'Fourth position'
},inplace=True)
data['Smoking'].replace({
    0:'Not a smoker',
    1:'Smoker',
    2:'Unknown'
},inplace=True)
data.head(10)
```

```python
#dropping columns
# in pandas column/vertical is axis-1, and row/horizontal is axis-0
data.drop(['MileMinDur'],axis=1,inplace=True)
data.info()
```

```python
#drop rows(duplicates)
data.drop_duplicates()
data.info()
```

```python
#drop rows on specifications
new_data =data[data['Gender'] != 'Male']
new_data.info()
```

```python
# removing date null cells using Timestamp
data['bday'] = data['bday'].fillna(pandas.Timestamp('2000-01-01'))
data['enrolldate'] = data['enrolldate'].fillna(pandas.Timestamp('2000-01-01'))
data['expgradate'] = data['expgradate'].fillna(pandas.Timestamp('2000-01-01'))

data.isnull().sum()
```

```python
# Find Weight mean
mean_weight = data['Weight'].mean()
mean_height = data['Height'].mean()
mean_sprint = data['Sprint'].mean()
mean_english = data['English'].mean()
mean_reading = data['Reading'].mean()
mean_math = data['Math'].mean()
mean_writing = data['Writing'].mean()
mean_commutetime = data['CommuteTime'].mean()
mean_sleeptime = data['SleepTime'].mean()
mean_studytime = data['StudyTime'].mean()



#fill the empty cells with th
data['Weight'].fillna(mean_weight, inplace=True)
data['Height'].fillna(mean_height, inplace=True)
data['Sprint'].fillna(mean_sprint, inplace=True)
data['English'].fillna(mean_english, inplace=True)
data['Reading'].fillna(mean_reading, inplace=True)
data['Math'].fillna(mean_math, inplace=True)
data['Writing'].fillna(mean_writing, inplace=True)
data['CommuteTime'].fillna(mean_commutetime, inplace=True)
data['StudyTime'].fillna(mean_studytime, inplace=True)
data['SleepTime'].fillna(mean_sleeptime, inplace=True)

data.isnull().sum()
```

```python
data = data.assign(Score = lambda new_data: data.Reading + data.Writing + data.Math + data.English)
data.head(5)
```

```python
data = data.assign(BMI = lambda BMI : data.Weight / data.Height**2)
data.head(5)
```

```python
#converting Weight Pound to KG
def convert(x):
    answer = 0.453592 * x
    return answer

data['Weight'] = data['Weight'].apply(convert)

#converting Height Inch to CM
def convert(x):
    answer = 0.0254 * x
    return answer

data['Height'] = data['Height'].apply(convert)
data.head(5)
```
