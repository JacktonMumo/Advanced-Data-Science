```python
import pandas
data =pandas.read_csv('cleanedDataset.csv',parse_dates=['bday','enrolldate','expgradate'])
data.head(10)
```

```python

# matplotlib is a package that helps in creating visual statistical graphs
#seaborn is built on top of matplotlib that provides a higher level interface for creating graphs 
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
# univariate = pie chart, bargraph, histogram
# bivariate = scatterplot, boxplot,  swarmplot
# multivariate = scatterplot, heatmap,....

#univariate: countplot(nameofcolumn, color, hue)
sns.countplot(data['HowCommute'],palette='Oranges')
plt.title('Distribution of commute status')
plt.xlabel('Number of Students')
plt.ylabel('How They Commute')
```

```python
#Bivariate: barplot()
import numpy
ax = sns.barplot(x='Gender', y='Math',data=data,palette='Reds',estimator=numpy.sum) #estimator gives a way to calculate sum,std,prod
for p in ax.patches:
    txt = str(p.get_height().round(1)) + ''
    txt_x = p.get_x()
    txt_y = p.get_height()
    ax.text(txt_x,txt_y,txt)
```

```python
#Multivariate
sns.violinplot(x='Gender',y='Math',hue='State',palette='magma',data=data)
```

```python
subset = data[['Height','Sprint','Math','Reading','Writing']] #quantitative
sns.heatmap(subset.corr(),annot=True,cmap='Blues')
```
